#### Results & Findings Explained
 Supervised modeling demonstrated that a meaningful predictive structure exists in the dataset, but the choice of model is critical for determining performance. The data was initially split at 80/20, resulting in a training set of 20,000 and a test set of 5,000. To prioritize the identification of at-risk students, who are a minority class, and align with the early identification objective, a weighting scheme was implemented to penalize misclassifying at-risk students more severely. The resulting logistic regression model showed no predictive value, with an ROC-AUC of 51% and an overall recall of 50%, which is no better than random chance. Similarly, the highest identified coefficient factor was only 0.013, suggesting minimal explanatory value for attendance decline. This suggests that the dataset lacks sufficient predictive signal to support early identification of at-risk students. However, to continue the evaluation and test the early result, random forest modeling was conducted.  
 Random forest modeling achieved substantially stronger results with an ROC-AUC of 75% and an overall recall value of 65%. This confirmed the presence of non-linear relationships among the predictors. Feature analysis revealed that daily study hours overwhelmingly dominate predictive power, which, while intuitive, reinforces the importance of behavioral engagement as more informative than demographic characteristics for at-risk identification. Threshold-tuning analysis was then conducted to refine the model's results as needed for operational purposes, based on institutional resources, desired outcomes, or policy thresholds. This analysis shows that recall values exceeding 75% were achievable at moderate at-risk probability thresholds (0.36), indicating that effective early-warning detection can occur without extreme over-classification, with 2680 students identified at this threshold. This strengthens the feasibility of deploying predictive analytics as a screening and support prioritization tool for educational institutions.  
#### Implementation and Ethical considerations
 While the Random Forest model demonstrated meaningful predictive capability, operationalizing its use must be carefully governed to prevent unintended harm. A primary concern is algorithmic bias. Even when sensitive demographic variables are excluded from modeling, proxy relationships may still encode structural bias. Factors such as study hours, travel time, or access to resources may correlate with socioeconomic conditions rather than student motivation alone. Fairness auditing would be required to ensure the model did not disproportionately flag already disadvantaged students and reinforce inequities, rather than mitigating them. Assessments across demographic subgroups regularly and monitoring of intervention outcomes would be required prior to real-world implementation.  
 Additionally, false positives and student labeling raise additional concerns. Threshold tuning demonstrated that improving recall – a priority for early intervention – requires increasing the number of students flagged for support. While acceptable in a supportive context, this misclassification could create stigma, unnecessary monitoring, or reduce student trust if predictions are interpreted as judgments rather than risk indicators. Model performance could also be further impacted, as many behavioral factors may be self-reported and, as a result, skewed to avoid perceived negative impacts. This reinforces the need to frame the outputs as strictly decision-support signals to guide voluntary outreach and resource allocation, not disciplinary or evaluative actions.  
 Transparency and explainability also present ethical obligations. Tree-based ensemble models, while more accurate than linear approaches in this effort, are less inherently interpretable. Clear communication about which data are being used, how predictions are generated, and how the results are used to inform interventions is required. This includes mitigating the risk of function creep in using this for other secondary uses, such as admissions screening.

#### Prepare dataset for modeling and additional exploration of prediction factors

```python
df_model = pd.read_csv('Student_Performance.csv')
df_model.head()
```

```python
# Removing student id column, no predictive value, and gender, excluding to reduce ethical risk and bias concerns.
# Removing subject individual scores and final grade as overall score was used for analysis and would lead to target leakage is they remained.

columns_to_drop = [
    'student_id',
    'gender',
    'attendance_percentage',
    'math_score',
    'science_score',
    'english_score',
    'overall_score',
    'final_grade'
]

df_model = df_model.drop(columns=columns_to_drop)
```

```python
df_model.head()
```

```python
# Define target variable: early attendance risk
df_model['low_attendance_risk'] = (df['attendance_percentage'] < 70).astype(int)

# Verify class distribution
df_model['low_attendance_risk'].value_counts(normalize=True)
```

Based on the current values associated with the travel_time column I will be treating them as ordinal and encoding them as the time constraint is ordinal and distance would logically increase burden which could contribute to lower attendance.

```python
# Travel time mapping
travel_time_mapping = {
    '<15 min': 0,
    '15-30 min': 1,
    '30-60 min': 2,
    '>60 min': 3
}

df_model['travel_time_encoded'] = df_model['travel_time'].map(travel_time_mapping)
df_model['travel_time_encoded'] = df_model['travel_time_encoded'].astype(int)
```

```python
df_model.head()
```

```python
df_model.isna().sum().sort_values(ascending=False)
```

```python
# Binary mappings
binary_mappings = {
    'internet_access': {'yes': 1, 'no': 0},
    'extra_activities': {'yes': 1, 'no': 0},
    'school_type': {'public': 0, 'private': 1}
}

for col, mapping in binary_mappings.items():
    if col in df_model.columns:
        df_model[col] = df_model[col].map(mapping)

binary_columns = [
    'internet_access',
    'extra_activities',
    'school_type'
]

df_model[binary_columns] = df_model[binary_columns].astype(int)
df_model.head()
```

```python
categorical_cols = ['parent_education', 'study_method']

# One-hot encoding for remaining categorical columns
df_model = pd.get_dummies(df_model, columns=categorical_cols, drop_first=True)
```

```python
df_model.info()
```

All variables are now properly encoded for use in logistic regression and decision tree models. Data types are correct with the exceptiuon that travel_time original column was was left unchanged with the encoded column as a separate identifier. Exploration of the additional factors to identify strenght of potential relationships with the target at risk variable will now be conducted.

```python
numeric_features = df_model.select_dtypes(include=['int64', 'int32', 'float64', 'bool']).columns
numeric_features
```

```python
correlation_with_target = (
    df_model[numeric_features]
    .corr()['low_attendance_risk']
    .sort_values(ascending=False)
)

correlation_with_target
```

Initial correlation value analysis shows that the correlations may be small which is expected for nonlinear relationships. These may still perform well in logistic and tree-based models. This is also to be expected when dealing with social or behavorial data as it can be very context sensitive. Feature selection will be determined by multivariate modeling rather than correlation magnitude alone.

The data will now be split into trianing and test sets that will also require stratification to preserve proportions. An 80/20 test split will be utilized. In order to keep the split evaluation consistent over the project a fixed random state will be used.

```python
from sklearn.model_selection import train_test_split

X = df_model.drop(columns=['low_attendance_risk'])

X = X.select_dtypes(include=['int64', 'float64', 'int32', 'float32'])

y = df_model['low_attendance_risk']

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    stratify=y,
    random_state=42
)

print("Train size:", X_train.shape)
print("Test size:", X_test.shape)
print("\nClass balance (train):")
print(y_train.value_counts(normalize=True))
```

As the basis for our modeling we want to prioritize the the identification of at risk students which will be the minority class. As a result, we will be weighting to penalize misclassification of at-risk students more than misclassification of other students.

```python
from sklearn.utils.class_weight import compute_class_weight
import numpy as np

classes = np.unique(y_train)
weights = compute_class_weight(
    class_weight="balanced",
    classes=classes,
    y=y_train
)

class_weights = dict(zip(classes, weights))
class_weights
```

Train and evaluate logistic regression model

```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(
    class_weight=class_weights,
    max_iter=1000,
    random_state=42
)

log_model.fit(X_train, y_train)

from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    roc_auc_score,
    RocCurveDisplay
)
import matplotlib.pyplot as plt

y_pred = log_model.predict(X_test)
y_prob = log_model.predict_proba(X_test)[:, 1]

print(classification_report(y_test, y_pred))

cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)

roc_auc = roc_auc_score(y_test, y_prob)
print("ROC-AUC:", roc_auc)

RocCurveDisplay.from_predictions(y_test, y_prob)
plt.show()
```

The logistic regression classifier achieved an ROC-AUC of approximately 0.51 and an overall accuracy of 50%, indicating performance equivalent to random chance. Confusion matrix analysis further demonstrated limited discriminative capability, with recall for the at-risk class of only 0.51 and precision of 0.40.

```python
import pandas as pd

coef_df = pd.DataFrame({
    "Feature": X.columns,
    "Coefficient": log_model.coef_[0]
}).sort_values(by="Coefficient", ascending=False)

coef_df
```

Coefficient magnitudes across all predictors were near zero, suggesting minimal explanatory power of the available variables for attendance decline. These findings indicate that the dataset lacks sufficient predictive signal to support reliable early identification of students at risk for low attendance. However, to ensure the is accurate this will be evaluated through Random Forest model which should be able to perform better any potential nonlinear relationships to determine if there is any real signal available in the data.

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(
    n_estimators=300,
    class_weight="balanced",
    random_state=42,
    n_jobs=-1
)

rf_model.fit(X_train, y_train)
```

```python
y_pred_rf = rf_model.predict(X_test)
y_prob_rf = rf_model.predict_proba(X_test)[:, 1]

print(classification_report(y_test, y_pred_rf))

cm_rf = confusion_matrix(y_test, y_pred_rf)
print("Confusion Matrix:\n", cm_rf)

roc_auc_rf = roc_auc_score(y_test, y_prob_rf)
print("ROC-AUC:", roc_auc_rf)

RocCurveDisplay.from_predictions(y_test, y_prob_rf)
plt.show()
```

```python
rf_importance = pd.DataFrame({
    "Feature": X.columns,
    "Importance": rf_model.feature_importances_
}).sort_values(by="Importance", ascending=False)

rf_importance
```

In opposition to the findings with linear regression, the results of the random forest model shows a substantial predictive signal indicating nonlinear relationships and feature interactions. The ROC-AUC of .75 indicates a clear improvement to random chance that was found with linear regression. The Recal of .65 indicates correct identification of 65% of at-risk students which is significant as an early warning as an indicator for intervention. However, there are still false positives for identification of at risk students with the precision score of .61.

It was identified that study hours is overwhelmingly the strongest predictor of at-risk students. While this may seem like common sense it is still important to identify and reinforces that behavorial engagement matters more in these factors than demographic factors.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import (
    precision_score,
    recall_score,
    f1_score,
    accuracy_score
)

# Predicted probabilities for the positive (at-risk) class
y_probs = rf_model.predict_proba(X_test)[:, 1]

thresholds = np.linspace(0.05, 0.95, 50)

results = []

for t in thresholds:
    y_pred_t = (y_probs >= t).astype(int)

    results.append({
        "threshold": t,
        "accuracy": accuracy_score(y_test, y_pred_t),
        "precision": precision_score(y_test, y_pred_t),
        "recall": recall_score(y_test, y_pred_t),
        "f1": f1_score(y_test, y_pred_t),
        "students_flagged": y_pred_t.sum()
    })

threshold_df = pd.DataFrame(results)
threshold_df.head()
```

Predicted class probabilities for the at-risk category were generated using the model’s probability output and used to evaluate classification performance across varying decision thresholds rather than relying on the default 0.50 cutoff.

```python
plt.figure()
plt.plot(threshold_df["threshold"], threshold_df["recall"], label="Recall")
plt.plot(threshold_df["threshold"], threshold_df["precision"], label="Precision")
plt.plot(threshold_df["threshold"], threshold_df["f1"], label="F1 Score")

plt.xlabel("Decision Threshold")
plt.ylabel("Score")
plt.title("Threshold Tuning for At-Risk Student Detection")
plt.legend()
plt.grid(True)
plt.show()
```

Plotting based on the threshold the number of student that would be identified as at risk to judge real impact of potential intervention opportunities.

```python
plt.figure()
plt.plot(threshold_df["threshold"], threshold_df["students_flagged"])

plt.xlabel("Decision Threshold")
plt.ylabel("Number of Students Flagged At-Risk")
plt.title("Operational Impact of Threshold Choice")
plt.grid(True)
plt.show()
```

Based on the desired outcome, policy thresholds, or institution resources we can then adjust to find the best threshold.

```python
# Threshold maximizing recall (early-warning focus)
best_recall_row = threshold_df.loc[threshold_df["recall"].idxmax()]
best_recall_row
```

```python
# Threshold maximizing F1 score (balanced performance)
best_f1_row = threshold_df.loc[threshold_df["f1"].idxmax()]
best_f1_row
```

```python
# Custom policy threshold to provide the most efficient intervention threshold
policy_threshold = (
    threshold_df[threshold_df["recall"] >= 0.75]
    .sort_values("precision", ascending=False)
    .iloc[0]
)

policy_threshold
```

This demonstrates the model is able to still provide high recall values even without blowing out the number of identified students. At threshold 0.34, the model detects 78% of at-risk students while flagging only 42% of the population.
